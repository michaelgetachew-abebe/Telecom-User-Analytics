### Handling Missing Values

It is a customary to calculate the amount of missing data in pur dataframe in percentage since it would enable us to either impute(insert) data or drop the missing values if they can be neglected.It is a recommended practice to drop the rows with missing values if the percentage is less than 10%

```python
#Perform missing percentage calcualation....
def percent_missing(df):

    # Calculate total number of cells in dataframe
    totalCells = np.product(df.shape)

    # Count number of missing values per column
    missingCount = df.isnull().sum()

    # Calculate total number of missing values
    totalMissing = missingCount.sum()

    # Calculate percentage of missing values
    print("The Diabetes dataset contains", round(((totalMissing/totalCells) * 100), 2), "%", "missing values.")

percent_missing(db)
```

```python
#Calculate the percent missing in each column so that those columns having the missing values greater than 30 could be removed
percent_missing_column = (db.isnull().sum() / len(db))*100
percent_missing_value_df = pd.DataFrame({'column_name': db.columns,
                                 'percent_missing': percent_missing_column})
percent_missing_value_df
```

```python
#From the scores of the missing percentages look for columns with percentage greater than 30%. 
drop_list = percent_missing_value_df[percent_missing_value_df.percent_missing > 30]
drop_list
```

```python
# drop columns with more than 30% missing values

df_clean = db.drop(['TCP DL Retrans. Vol (Bytes)', 'TCP UL Retrans. Vol (Bytes)', 'HTTP DL (Bytes)', 'HTTP UL (Bytes)', 'Nb of sec with 125000B < Vol DL','Nb of sec with 1250B < Vol UL < 6250B','Nb of sec with 31250B < Vol DL < 125000B','Nb of sec with 37500B < Vol UL','Nb of sec with 6250B < Vol DL < 31250B','Nb of sec with 6250B < Vol UL < 37500B'], axis=1)
df_clean.shape
```

```python
# fill missing with ffill method for columns (diag_1, diag_2, diag_3)

def fix_missing_ffill(df, col):
    df[col] = df[col].fillna(method='ffill')
    return df[col]


def fix_missing_bfill(df, col):
    df[col] = df[col].fillna(method='bfill')
    return df[col]

df_clean['diag_1'] = fix_missing_ffill(df_clean, 'diag_1')
df_clean['diag_2'] = fix_missing_ffill(df_clean, 'diag_2')
df_clean['diag_3'] = fix_missing_ffill(df_clean, 'diag_3')

# fill 'race' column with mode 
df_clean['race'] = df_clean['race'].fillna(df_clean['race'].mode()[0])
```
